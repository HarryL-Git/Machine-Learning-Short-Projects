## Advanced Techniques:
* Bagging (Bootstrap AGGregatING)

```python
# import library for decison tree and data 
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier

# load the data
wine = load_wine()

# select specific data
X, y = wine.data[:, 11:13], wine.target
```

```python
# run a decison tree
clf = DecisionTreeClassifier()
clf.fit(X, y)
```

```python
# plot decision tree score
clf.score(X, y)
```

```python
# import library for bagging
from sklearn.ensemble import BaggingClassifier

# set a bagging classifier
bg = BaggingClassifier(base_estimator = clf, max_samples = 0.1, max_features = 0.1, n_estimators = 10)
```

```python
# fit the classifier
bg.fit(X, y)
```

```python
# find the score of the classifier
bg.score(X, y)
```

```python
# import library for result ploting
from mlxtend.plotting import plot_decision_regions

# plot the decision regions
plot_decision_regions(X, y, bg)
```

# Task Three

```python
# import libary for Adaboosting
from sklearn.ensemble import AdaBoostClassifier
import matplotlib.pyplot as plt

# create a decision tree
clf = DecisionTreeClassifier(criterion = 'entropy', max_depth = 1)
n_est = [1, 2, 5, 10]

# create an Adaboosting model based on the decision tree
for num_est in n_est:
  clf_boost = AdaBoostClassifier(clf, n_estimators = 1)

  # fit the model
  clf_boost.fit(X, y)

  # get the score of the model
  print(clf_boost.score(X, y))

  # plot the results
  plot_decision_regions(X, y, clf_boost)
  plt.show()
```

# Task Four

## Boosting algorithms:

* AdaBoost : Adaptive boosting

```python
# import libraries
from sklearn.model_selection import train_test_split

# split data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# import library for Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier

# train the model
model = GradientBoostingClassifier()
model.fit(x_train, y_train)

# get the model score
model.score(x_test, y_test)
```

```python
# import library for XGBM
import xgboost as xgb

# train the model
model = xgb.XGBClassifier()
model.fit(x_train, y_train)

# get the model score
model.score(x_test, y_test)
```

# Task Five

## Ensemble Learning:
* Homogeneous ensemble methods : Bagging , Boosting
* Heterogeneous ensemble methods: Stacking

```python
# import libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import six
import sys
sys.modules['sklearn.externals.six'] = six
from mlxtend.classifier import StackingClassifier
```

```python
# creat the models
clfk = KNeighborsClassifier(n_neighbors = 1)
clfg = GaussianNB()
clfr = RandomForestClassifier()
lr = LogisticRegression()

# create the stacking model
sclf = StackingClassifier(classifiers = [clfk, clfg, clfr], meta_classifier=lr)
```

```python
clfs = [clfk, clfg, clfr, sclf]

# create a classifier model
for classifier in clfs:
  # fit the model
  classifier.fit(x_train, y_train)

  # print the results
  print('Acc', classifier.score(x_test, y_test))

  # plot the results
  plot_decision_regions(x_test, y_test, classifier)
  plt.show()
```
